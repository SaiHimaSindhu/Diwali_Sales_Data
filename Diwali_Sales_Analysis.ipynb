# Exploratory Data Analysis

```python
df['Product_Category'].unique()
```

```python
region_product = df.groupby(['Zone', 'Product_Category'])['Amount'].sum().reset_index()

print(region_product)
```

```python
plt.figure(figsize=(20,7))

sns.barplot(x='Zone',y='Amount',hue='Product_Category',data=region_product)

plt.xticks(rotation=90)

plt.title("Zone Wise Product Sales Analysis")

plt.show()
```

```python
product_orders = df.groupby('Product_Category')['Orders'].sum().reset_index()

print(product_orders)
```

```python
plt.figure(figsize=(14,6))

sns.barplot(x='Product_Category',y='Orders',data=product_orders)

plt.xticks(rotation=90)

plt.title("Highest Ordered Products")

plt.show()
```

```python
age_data = df.groupby('Age Group')['Amount'].sum().reset_index()

print(age_data)
```

```python
plt.figure(figsize=(10,5))

sns.barplot(x='Age Group',y='Amount',data=age_data)

plt.title("Age Group vs Purchase Amount")

plt.show()
```

```python
gender_product = df.groupby(['Gender', 'Product_Category'])['Orders'].sum().reset_index()

print(gender_product)
```

```python
plt.figure(figsize=(14,6))

sns.barplot(x='Product_Category',y='Orders',hue='Gender',data=gender_product)

plt.xticks(rotation=90)

plt.title("Gender Wise Product Orders")

plt.show()
```

```python
product_amount = df.groupby('Product_Category')['Amount'].sum().reset_index()

print(product_amount)
```

```python
plt.figure(figsize=(14,6))

sns.barplot(x='Product_Category',y='Amount',data=product_amount)

plt.xticks(rotation=90)

plt.title("Highest Purchase Amount By Product")

plt.show()
```

```python
zone_avg = df.groupby('Zone')['Amount'].mean().reset_index()

print(zone_avg)
```

```python
plt.figure(figsize=(10,5))

sns.barplot(x='Zone',y='Amount',data=zone_avg)

plt.title("Average Purchase Amount From Every Zone")

plt.show()
```

```python
marital_product = df.groupby(['Marital_Status', 'Product_Category'])['Amount'].sum().reset_index()

print(marital_product)
```

```python
plt.figure(figsize=(14,6))

sns.barplot(x='Product_Category',y='Amount',hue='Marital_Status',data=marital_product)

plt.xticks(rotation=90)

plt.title("Product Purchase Based On Marital Status")

plt.show()
```
